## Умножение матриц

В этом задании вы можете записывать формулы в *Jupyter Notebook.*

Чтобы записать формулу внутри текста, окружите её символами доллара \\$; если снаружи —  двойными символами \\$\\$. Эти формулы записываются на языке вёрстки *LaTeX.* 

Для примера мы записали формулы линейной регрессии. Можете их скопировать и отредактировать, чтобы решить задачу.

Работать в *LaTeX* необязательно.

Обозначения:

- $X$ — матрица признаков (нулевой столбец состоит из единиц)

- $y$ — вектор целевого признака

- $P$ — матрица, на которую умножаются признаки

- $w$ — вектор весов линейной регрессии (нулевой элемент равен сдвигу)

Предсказания:

$$
a = Xw
$$

Задача обучения:

$$
w = \arg\min_w MSE(Xw, y)
$$

Формула обучения:

$$
w = (X^T X)^{-1} X^T y
$$

Вычислим связь между $w$ и $w_p$. При помощи замен запишем:
$$
w=(X^TX)^{-1}X^Ty=AX^Ty=By
$$
Тогда для $w_p$ можем записать:
$$
w_p=((XP)^T(XP))^{-1}(XP)^Ty=(P^TX^TXP)^{-1}P^TX^Ty=(P^TAP)^{-1}P^TX^Ty=P^{-1}A^{-1}(P^T)^{-1}P^TX^Ty=P^{-1}A^{-1}EX^Ty=P^{-1}A^{-1}X^Ty=P^{-1}By
$$
Соответственно получаем:
$$
w_p=P^{-1}w
$$

Воспользуемся упрощённой формулой вычисления предсказаний:
$$
a=Xw
$$
Для изменённой матрицы она выглядит следующим образом:
$$
a'=XPw_p
$$
Подставляя полученную выше формулу связи $w$ и $w_p$ получаем:
$$
a'=XPP^{-1}w=XEw=Xw=a
$$
Из чего следует, что задачи вычисления предсказаний для матриц $X$ и $X_p$ должны привести к одинаковым предсказаниям.

### Вычислим значения R2 и MSE по исходным данным до их изменения

Предположим, что при умножении значений на обратимую матрицу, размер которой равен количству признаков в изначальном датасете модель линейной регрессии сможет предсказать значения аналогично. Обратимая матрица требуется для последующего восстановления данных путём умножения на обратную матрицу.

Создадим обратимую матрицу:

```python
matrix=np.random.normal(3, size=(4,4))
matrix
```

Проверим матрицу на обратимость

```python
np.linalg.inv(matrix)
```

Матрица обратима.

Умножим значения, содержащиеся в датафрейме `features` на `matrix`:

```python
train_features_matrix=train_features.dot(matrix)
train_features_matrix.columns = train_features.columns
train_features_matrix.head()
```

```python
test_features_matrix = test_features.dot(matrix)
test_features_matrix.columns = test_features.columns
test_features_matrix.head()
```

### Ответ

При умножении исходных данных на обратимую матрицу качество линеёной регрессии **остаётся неизменным**.

### Пояснение

Изменение качества предсказания линейной регрессии не происходит при умножении признаков исходных данных на обратимую матрицу, т.к. параметры линейной регрессии в исходной задаче и в преобразованной связаны между собой операцией матричного умножения, в результате которой Изменяются лишь числовые параметры самого датасета, при этом в каждом столбце происходит умножение по одинаковой формуле. В результате чего на обучении модель лишь подберёт другие числовые параметры, которые приведёт к аналогичным предсказаниям. В результате получаем преобразованные в числовом виде данные, которые не влияют на качество предсказания.

Таким образом, если изменить исходные данные методом умножения признаков на обратимую матрицу, не не изменив значений целевого признака, качество предсказания модели не изменится.

## Алгоритм преобразования

### Описание алгоритма

1. Создадим случайную матрицу $P$;
2. Проверим её на обратимость;

4. Найдём метрика качества `R2_score`, `MSE` на исходных и преобразованных данных;
5. Изучим полученный результат и напишем вывод.

### Обоснование

Если мы произведём умножение признаков тренировочной и тестовой выборок на одну и ту же обратимую матрицу, то результатом предсказания будет произведение нашей изменённой таблицы с признаками на вектор, который был получен из той же таблицы с изменёнными данными увеличенную на ту же величину, что и признаки тренировочных данных. Проверим работу данного алгоритма.

## Проверка алгоритма

Создадим квадратную матрицу $P$ такую, чтобы она была обратимой. Создадим функцию, которая будет генерировать матрицу $P$ с размерами равными количеству признаков в данных до того момента, пока определитель матрицы будет не нулевым.

```python
def get_rand_matrix(data):
    det = 0
    while det == 0:
        matrix = np.random.normal(size=(data.shape[1], data.shape[1]))
        det = np.linalg.det(matrix)
    return matrix
```

```python
P = get_rand_matrix(train_features)
P
```

Матрица обратима. Умножим признаки тренировочной и тестовой выборок на новую матрицу. Полученные матрицы будем обозначать при помощи `..._mat`.

```python
train_mat = train_features.dot(P)
train_mat.columns = train_features.columns
train_mat.head()
```

```python
test_mat = test_features.dot(P)
test_mat.columns = test_features.columns
test_mat.head()
```

Целевые признаки изменять не будем.

Добавим в полученную матрицу единичный столбец на первое место.

```python
X = np.concatenate((np.ones((train_mat.shape[0], 1)), train_mat), axis=1)
```

```python
X = pd.DataFrame(X)
X.head()
```

Присвоим переменной $у$, целевые признаки и вычислим по формуле ( $w = (X^T X)^{-1} X^T y$ )  вектор $w$

```python
y = train_target
w = (np.linalg.inv(X.T.dot(X)).dot(X.T)).dot(y)
```

Вычислим предсказание модели методом умножения тестовой выбрки $test_m$  и увеличим на величину сдвига значение $w0$  Переменную предсказанных значений обозначим $a$

```python
a = test_mat.dot(w[1:]) + w[0]
```

Выичслим новые значения `R2_score` и `MSE`

```python
print(f'R2_score: {r2_score(test_target, a).round(3)}')
print(f'MSE: {mean_squared_error(a, test_target).round(3)}')
```

### Графическое изображение предсказанных значений относительно истиных значений целевого столбца и сравним эти значения с предсказаниями по неизменённым признакам

```python
plt.figure(figsize=(9,9))
plt.plot(test_target, a, 'yo', alpha=0.7)
plt.plot(test_target, predictions, 'o', alpha=0.4)
plt.legend(['Предсказание модели нашего алгоритма c изменёнными признаками', 'Предсказание модели Sklearn c базовыми признаками'])
plt.title('График значений истиные/предсказанные')
plt.grid(True)
plt.xlabel('Истинные значения целевого признака')
plt.ylabel('Предсказанные значения')
plt.show()
```

Соберём в сводную таблицу полученные значения `R2_score` и `MSE` для изменённой и неизменённой таблиц признаков

```python
comparison_tabl = pd.DataFrame(index=['R2_score','MSE'], columns=['Изменённая матрица','Неизменённая матрица'])
comparison_tabl['Изменённая матрица'] = r2_score(test_target, a).round(5), mean_squared_error(a, test_target).round(5)
comparison_tabl['Неизменённая матрица'] = r2_score(test_target, predictions).round(5), mean_squared_error(predictions, test_target).round(5)

comparison_tabl.T
```
