### Как уже упомяналось на этапе изучения изначальных данных, данные несбалансированы. Проведём сначала обучение моделей на несбалансированных данных и посмотрим какие результаты сможем получить.

Будем обучать 3 различных модели - Решающее дерево `DecisionTreeClassifier`, Случайный лес `RandomForestClassifier` и логистическую регрессию `LogisticRegression`.

### Модели

#### Решающее дерево

Начнём с решающего дерева.
Поступим следующим образом:
* Будем задавать гиперпараметр `max_depth` в цикле от 1 до 15 последовательно.
* На каждом проходе цикла будем проверять модель на валидационной выборке по метрикам `accuracy` и `F1`. Наилучшую модель по каждой из метрик сохраним в соответствующей переменной.

```python
acc_tree = 0
acc_tree_best = 0
f1_tree = 0
f1_tree_best = 0
best_depth_acc = 0
best_depth_f1 = 0
depth_list = [*range(1,16)]
acc_tree_list = []
f1_tree_list = []
for depth in range(1,16):
    model_tree = DecisionTreeClassifier(max_depth=depth, random_state=12345)
    model_tree.fit(features_train, target_train)
    predictions_tree = model_tree.predict(features_valid)
    acc_tree = accuracy_score(predictions_tree, target_valid)
    acc_tree_list.append(acc_tree)
    if acc_tree > acc_tree_best:
        model_tree_best_acc = model_tree
        best_depth_acc = depth
        acc_tree_best = acc_tree
    f1_tree = f1_score(target_valid, predictions_tree)
    f1_tree_list.append(f1_tree)
    if f1_tree > f1_tree_best:
        model_tree_best_f1 = model_tree
        best_depth_f1 = depth
        f1_tree_best = f1_tree
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(depth_list, acc_tree_list, label='Accuracy')
ax.plot(depth_list, f1_tree_list, label='F1-score')
ax.legend()
plt.show()
probabilities_valid = model_tree_best_f1.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]
fpr, tpr, thresholds = roc_curve(target_valid, probabilities_one_valid)
plt.figure(figsize=(15,15))
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая')
plt.show()
print(f'Наилучшая точность модели: {acc_tree_best:.1%} '
    f'была достигнута при глубине решающего дерева {best_depth_acc}\n'
    f'Наилучшая F1-мера модели: {f1_tree_best:.4} '
    f'была достигнута при глубине решающего дерева {best_depth_f1}\n'
    f'AUC-ROC для модели с лучшей F1-мерой: {roc_auc_score(target_valid, probabilities_one_valid):.4}')
```

Получаем следующие наилучшие модели по метрикам `Accuracy` и `F1`:
 - Наилучшую Accuracy, равную 85,9% показала модель  глубиной решающего дерева равной 6;
 - Наилучшую F1-меру, равную 0,5804 показала модель с глубиной решающего дерева, равной 7.

При глубине решающего дерева 6 или 7 модель скорее всего начинает переобучаться и `Accuracy` снижается.

#### Случайный лес

Теперь построимм модель на основе случайного леса. У случайноо леса также есть несколько гиперпараметров, которые можно изменять: `n_estimators` - количество решающих деревьев леса, `max_depth` - максимальная глубина решающего дерева. Поступим следующим образом:

* Создадим цикл, в котором будем изменять количество решающих деревьев от 10 до 200 с шагом в 10.
* Создадим вложенный цикл, в котором будем изменять максимальную глубину решающего дерева от 1 до 15 с шагом в 1.
* Во вложенном цикле будем создавать модель, тренировать и проверять на валидационной выборке. Если точность модели выше, чем максимальная достигнутая до этого точность, будем записывать модель, глубину решающего дерева и количество деревьев в специальные переменные.
* В конце выведем графики, на которых покажем точность моделей в зависимости от параметров.

Также будем испоьзовать гиперпараметр `random_state=12345` для воспроизводимости результатов.

## Тестирование модели

Наилучшей с точки зрения F1-меры получилась модель случайного леса, которая была получена при балансировке увеличением выборки. Проверим её на тестовой выборке и, если F1-мера превысит 0,59, построим для неё параметры полноты и точности, матрицу ошибок, а также ROC-меруи кривую.

```python
print(f'{f1_score(target_test, model_best_upsampled.predict(features_test)):.4}')
```

Получаем, что модель преодолела требуемый порог в 0,59. Получим для неё остальные метрики и построим ROC кривую.

```python
predicted_best_test = model_best_upsampled.predict(features_test)
probabilities_test = model_best_upsampled.predict_proba(features_test)
probabilities_one_test = probabilities_test[:, 1]
fpr, tpr, thresholds = roc_curve(target_test, probabilities_one_test)
print(
    f'Accuracy: {model_best_upsampled.score(features_test, target_test):.1%}\n'
    f'Recall: {recall_score(target_test, predicted_best_test):.4}\n'
    f'Precision: {precision_score(target_test, predicted_best_test):.4}\n'
    f'Матрица ошибок:\n {confusion_matrix(target_test, predicted_best_test)}\n'
    f'AUC-ROC: {roc_auc_score(target_test, probabilities_one_test):.4}')
plt.figure(figsize=(15,15))
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая')
plt.show()
```
